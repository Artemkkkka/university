```python
import numpy as np
import cv2 as cv 
import matplotlib.pyplot as plt
```

```python
filename = "img.png"
image = cv.imread(filename)
```

```python
plt.imshow(image)
plt.show
```

```python
image
```

```python
def isbw(A):
    if(len(A.shape)<3):
        for i in range(A.shape(0)):
            for j in range(A.shape(1)):
                if (isinstance(A(i,j), bool)):
                    return 1
    else: 
        return 0
```

```python
isbw(image)
```

```python
def isgray(A):
    if(len(A.shape)<3):
        return 1
    else: 
        return 0
```

```python
isgray(image)
```

```python
def isrgb(A):
    if(len(A.shape)==3):
        return 1
    else: 
        return 0
```

```python
isrgb(image)
```

```python
I0 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(I0)
plt.show
```

```python
# RGB-   
image_gray = cv.cvtColor(I0, cv.COLOR_BGR2GRAY)

#   RGB- 
image_rgb1 = cv.cvtColor(image_gray, cv.COLOR_GRAY2RGB)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(16, 11))
ax[0].imshow(I0) #  
print(isgray(image_gray))
ax[1].imshow(image_gray, cmap="gray") #  
print(isrgb(image_rgb1))
ax[2].imshow(image_rgb1) #  RGB-
```
